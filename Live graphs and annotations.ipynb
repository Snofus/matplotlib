```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
from matplotlib import style
style.use('seaborn')

```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

def animate(i):
    graph_data = open('Example.txt','r').read()
    lines = graph_data.split('\n') 
    xs = []
    ys =[]
    for line in lines:
        if len(line) > 1:
            x,y = line.split(',')
            xs.append(x)
            ys.append(y)
    ax1.clear()
    ax1.plot(xs,ys)
    bbox_props = dict(boxstyle = 'round',fc ='w',ec = 'k') # the squared annotate format
    ax1.annotate('Bad News',xy = (5,5),xytext=(6,6),arrowprops = dict(facecolor = 'grey', edgecolor = 'green'),bbox= bbox_props)

    font_dict = {'family':'serif',  # change the feature of the font
                'color':'darkred',
                'size': 15}
    ax1.text(0,1,'This is where we\ngonna start baby',fontdict = font_dict) #adding text to the graph
    
ani = animation.FuncAnimation(fig,animate,interval = 1000) # make the graph updated in every 1000ms
plt.show()                      

                      
            
                      
    
        
                      
```
