```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [5,7,4]

x2 = [1,2,3]
y2 = [10,14,12]

# label the lines
plt.plot(x,y,label = 'First Line')
plt.plot(x2,y2,label = 'Second Line')

# label the x and y axis
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph')
plt.grid(axis = 'y')
plt.legend()
```
