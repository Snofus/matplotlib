```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
fig = plt.figure(figsize=(14,8))
ax1 = plt.subplot2grid((2,3),(0,0),colspan=3)
ax2 = plt.subplot2grid((2,3),(1,0),colspan=1)
ax3 = plt.subplot2grid((2,3),(1,1),colspan=2)

def create_data():
    x = []
    y = []
    for i in range(10):
        x.append(i)
        y.append(random.randint(0,20))
    return x, y

x,y = create_data()
ax1.plot(x,y)

x,y = create_data()
ax2.plot(x,y)

x,y = create_data()
ax3.plot(x,y)
```

```python
def graph_data(stock,start="2018-6-16",end = "2019-6-16",ma1=5,ma2=30): #automatically download data from yahoo finance and plot the daily closing price
    import pandas as pd
    import matplotlib.pyplot as plt
    import pandas_datareader as wb
    import numpy as np
    import matplotlib.ticker as mticker # import to set the number of ticker on axis
    import matplotlib.dates as mdates
    pd = wb.DataReader(stock, "yahoo", start = start,end = end)
    fig = plt.figure(figsize=(14,8)) # define figure
    ax1 = plt.subplot2grid((5,1),(0,0),rowspan=4) # define subplot axis
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%y-%m-%d')) # As the x-axis is shared, ax1 needs to be adjusted as well
    ax1.xaxis.set_major_locator(mticker.MaxNLocator(15))
    plt.title(f'{stock}')
    ax1.set_ylabel('Price')
    ax2 = plt.subplot2grid((5,1),(4,0),sharex = ax1) #share the x-axis with subplot 1
    ax2.set_xlabel('Date') # to adjust the label and legend for each subplot,plt function needs to be called right after the plot
    ax2.set_ylabel('Volume')
    ax2.xaxis.set_major_locator(mticker.MaxNLocator(15))
    ax2.xaxis.set_major_formatter(mdates.DateFormatter('%y-%m-%d')) #set up the format of x-axis
#     ax2.yaxis.set_yticks(ticks = [1_000_000,10_000_000])
    
    
    ax1.plot(pd['Adj Close'],color = 'm',label = f'{stock}') 
    ax1.plot(pd['Adj Close'].rolling(ma1).mean(),label=f'{ma1} days moving average')
    ax1.plot(pd['Adj Close'].rolling(ma2).mean(),label=f'{ma2} days moving average')
    ax1.legend() # set up the legend for each subplot
    
    ax2.bar(pd.index,pd['Volume'],label = f'Daily trading volume')
    ax2.legend()
    
    ax1.fill_between(pd.index,pd['Adj Close'],pd['Adj Close'][0],where = (pd['Adj Close']<pd['Adj Close'][0]),alpha = 0.3,color = 'g') #solid graph rather than a line, the bottom line and the top line can be adjusted
    ax1.fill_between(pd.index,pd['Adj Close'],pd['Adj Close'][0],where = (pd['Adj Close']>pd['Adj Close'][0]),alpha = 0.3,color = 'r')
    ax1.grid(True,color = "g", linestyle = "dotted") # define subplot grid

    ax1.xaxis.label.set_color('c') #set the color of subplot x-axis label
    ax1.yaxis.label.set_color('r')
    ax1.set_yticks(np.linspace(min(pd['Adj Close']),max(pd['Adj Close']),15,dtype = int)) #set the y-ticker for the subplot
    for label in ax2.xaxis.get_ticklabels(): # for all the tick on the x-axis turn 45 degree
        label.set_rotation(45)

    plt.subplots_adjust(left = 0.09,bottom = 0.18,right = 0.94,top = 0.95,wspace = 0.2,hspace = 0)
```

```python
graph_data('TSLA')
```
