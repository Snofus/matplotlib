```python
import pandas as pd
import pandas_datareader as wb
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from mpl_finance import candlestick_ochl
import matplotlib.ticker as mticker
```

```python
def full_version_graph(stock,start="2019-6-16",end = "2020-6-16",ma1 = 5, ma2 = 20):
    # Collecting and processing data
    data = wb.DataReader(stock,'yahoo',start = start, end = end)
    candle_data = data[['Open','Close','High','Low']]
    candle_data = candle_data.reset_index()
    candle_data['Date'] = candle_data['Date'].map(mdates.date2num)
    
    # Label and ticker setting
    fig = plt.figure(figsize = (16,12))
    
    ax1 = plt.subplot2grid((4,1),(0,0),rowspan=1)
    ax1.plot(data['High']-data['Low'],'m',label = 'HL Difference')
    ax1.legend()
    leg1 = ax1.legend(loc = 2,ncol = 0,prop={"size":11}) # set the location and property of the legend
    leg1.get_frame().set_alpha = 0.4 # set the alpha of the frame of the legend
    ax1.xaxis.set_major_locator(mticker.MaxNLocator(15,prune = 'lower'))
    ax1.set_ylabel('HL Difference')
    

    
    ax2 = plt.subplot2grid((4,1),(1,0),rowspan=2,sharex = ax1)
    ax2.xaxis_date()
    ax2.xaxis.set_major_locator(mticker.MaxNLocator(15,prune = 'upper'))
    ax2.set_ylabel('Share Price')
    ax2.set_ylim(100,data['Adj Close'].max()+50) # scale the first y axis

    #second y-axis of ax2
    ax2v = ax2.twinx()
    ax2v.fill_between(data.index,data['Volume'],alpha =0.3,label = 'Trading Volume')
    ax2v.legend()
    leg2 = ax2v.legend(loc = 2,ncol = 0,prop={"size":11})
    ax2v.axes.yaxis.set_ticks([]) # hide the ticker of second y-axis
    ax2v.set_ylim(0,3*data['Volume'].max()) # scale the second y axis
    
    ax3 = plt.subplot2grid((4,1),(3,0),sharex = ax1,rowspan=1)
    ax3.xaxis.set_major_locator(mticker.MaxNLocator(15))
    ax3.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    for label in ax3.xaxis.get_ticklabels():
        label.set_rotation(45)
    ax3.set_xlabel('Date')
    ax3.set_ylabel('Moving Average')
    
    #Let the drawing begin
    candlestick_ochl(ax2,candle_data.values,colorup='r',colordown='g',width=0.5,)
    
    
    
    ax3.plot(data['Adj Close'].rolling(ma1).mean(),color = 'r',label = f'{ma1} days MA',linewidth =1)
    ax3.plot(data['Adj Close'].rolling(ma2).mean(),color = 'g',label = f'{ma2} days MA',linewidth =1)
    ax3.legend()
    leg3 = ax3.legend(loc = 2,ncol = 1,prop={"size":11})
    ax3.fill_between(data.index,data['Adj Close'].rolling(ma1).mean(),data['Adj Close'].rolling(ma2).mean(),
                     where =(data['Adj Close'].rolling(ma1).mean()>data['Adj Close'].rolling(ma2).mean()),
                    alpha = 0.3,color = 'r')
    ax3.fill_between(data.index,data['Adj Close'].rolling(ma1).mean(),data['Adj Close'].rolling(ma2).mean(),
                     where =(data['Adj Close'].rolling(ma1).mean()<data['Adj Close'].rolling(ma2).mean()),
                    alpha = 0.3,color = 'g')
    
    plt.subplots_adjust(left = 0.09,bottom = 0.18,right = 0.94,top = 0.95,wspace = 0.2,hspace = 0) # adjust the horizontal space between axes
    plt.setp(ax1.get_xticklabels(),visible = False) # invisible the x-label of ax1 and ax2
    plt.setp(ax2.get_xticklabels(),visible = False)
```

```python
full_version_graph('TSLA')
```
