```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (16,8))
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

slices = [7,2,7,8]
activities = ['sleeping','eating','working','playing']
plt.title('Pie Chart')

plt.pie(slices, labels = activities,startangle=90,explode=(0,0.1,0,0),autopct='%1.2f%%')
```
