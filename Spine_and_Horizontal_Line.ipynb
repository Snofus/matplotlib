```python
def graph_data(stock,start="2018-6-16",end = "2019-6-16"): #automatically download data from yahoo finance and plot the daily closing price
    import pandas as pd
    import matplotlib.pyplot as plt
    import pandas_datareader as wb
    import numpy as np
    pd = wb.DataReader(stock, "yahoo", start = start,end = end)
    fig = plt.figure(figsize=(14,8)) # define figure
    ax1 = plt.subplot2grid((1,1),(0,0)) # define subplot axis
    ax1.plot(pd['Adj Close'],label = f'{stock}') 
    ax1.axhline(pd['Adj Close'][0],color = 'k',linewidth = 1,linestyle = 'dotted') # the horizontal line
    ax1.axvline(pd.index[0],color = 'k',linewidth = 1) # the vertial line
#     ax1.fill_between(pd.index,pd['Adj Close'],pd['Adj Close'][0],where = (pd['Adj Close']<pd['Adj Close'][0]),alpha = 0.3,color = 'g') #solid graph rather than a line, the bottom line and the top line can be adjusted
#     ax1.fill_between(pd.index,pd['Adj Close'],pd['Adj Close'][0],where = (pd['Adj Close']>pd['Adj Close'][0]),alpha = 0.3,color = 'r')
    ax1.grid(True,color = "g", linestyle = "dotted") # define subplot grid
    ax1.spines['left'].set_color('k') # set the color the frame lines
    ax1.spines['right'].set_visible(False) # set the visibility of the frame lines
    ax1.spines['top'].set_visible(True)
    ax1.spines['top'].set_linewidth(5)
    ax1.tick_params(axis = 'y',colors = 'r') # set the color of the ticks
#     ax1.xaxis.label.set_color('c') #set the color of subplot x-axis label
#     ax1.yaxis.label.set_color('r')
    ax1.set_yticks(np.linspace(min(pd['Adj Close']),max(pd['Adj Close']),15,dtype = int)) #set the y-ticker for the subplot
    for label in ax1.xaxis.get_ticklabels(): # for all the tick on the x-axis turn 45 degree
        label.set_rotation(45)
    plt.legend()
    plt.subplots_adjust(left = 0.09,bottom = 0.18,right = 0.94,top = 0.95,wspace = 0.2,hspace = 0)
    plt.title("Stock_price_plot")
    plt.xlabel("Date")
    plt.ylabel("Share_Price $")
```

```python
graph_data('TSLA')
```
