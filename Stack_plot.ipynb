```python
import matplotlib.pyplot as plt
days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playinig = [8,5,7,8,13]

plt.stackplot(days,#x-axis
              sleeping,eating,working,playinig,colors = ['c','m','r','k'],labels = ('sleeping','eating','working','playing'))

plt.xlabel("x")
plt.ylabel('y')
plt.title('Interesting Graph')
plt.legend()
```
