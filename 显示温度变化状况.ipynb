```python
import matplotlib.pyplot as plt
import random
from pylab import mpl
```

```python
# 在python脚本中动态设置matplotlib字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
```

```python
# 数据生成
x = range(60)
y_sh = [random.uniform(10,20) for i in x]
y_bj = [random.uniform(21,35) for i in x]

# 创建画布
plt.figure(figsize = (12,6))

# 图形绘制
# 使用多次Plot可以在同一个axes中画多个折线
plt.plot(x,y_sh,label = '上海')
plt.plot(x,y_bj,label = '北京')

# 绘制x，y轴刻度
y_ticks = range(40)
x_ticks = range(60)
# 改变轴线标示
x_ticks_lable = [f'11:{i}' for i in x]
plt.yticks(y_ticks[::5])

# 注意xticks的首个参数必须是数字，之后的标示可以传入字符串进行修改
plt.xticks(x_ticks[::5],x_ticks_lable[::5])

# 添加网格
plt.grid(True,linestyle='--',alpha = 0.5)

# 添加描述
plt.xlabel('时间')
plt.ylabel('温度')
plt.title('中午11点至12点某城市温度变化图',fontsize = 15)
# 展现图例
plt.legend(loc='best')
# 储存图片
plt.savefig('Temp')
# 展示图片
plt.show()
```
