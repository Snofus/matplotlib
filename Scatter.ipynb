```python
import matplotlib.pyplot as plt
```

```python
# 1. 创建画布, 对于画布的设置需对figure传参
plt.figure(figsize = (20,8),dpi = 100)
x = [1,2,3,4,5,6,7]
y = [5,2,4,6,3,1,4]

# s代表size
plt.scatter(x,y,marker='*',s = 1000)
# savefig需要在show之前
plt.savefig('nihao')

# show会释放资源，即在show之后内存不再保存该图片
plt.show()
```

```python
plt.plot(x,y,marker='^',ms=10)
plt.grid()
```
